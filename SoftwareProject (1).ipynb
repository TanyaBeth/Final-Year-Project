```
#Function for ej to tonne conversion
def convert(input_in_ej):
  #output_in_tonnes
  #write the formula for tonne to ej conversion here
  output_in_tonnes = input_in_ej / 239005736.13767
  return output_in_tonnes

#Applying conversion to column
co2_tonne=[]
share_global_co2_tonne = []
co2_per_gdp_tonne = []

for index,row in carbondf.iterrows():
  keep0 = convert(row['co2'])
  keep1 = convert(row['share_global_co2'])
  keep2 = convert(row['co2_per_gdp'])

  co2_tonne.append(keep0)
  share_global_co2_tonne.append(keep1)
  co2_per_gdp_tonne.append(keep2)

df_tonne = pd.DataFrame(list(zip(co2_tonne, share_global_co2_tonne, co2_per_gdp_tonne)), 
               columns =['co2','share_global_co2','co2_per_gdp'])
print(df_tonne)
```
